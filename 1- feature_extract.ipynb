```python
import pandas as pd
import re
```

# Import dataset

```python
# download dataset from https://www.kaggle.com/datasets/shuyangli94/food-com-recipes-and-user-interactions/data?select=RAW_recipes.csv
# put dataset files inside a new folder named as 'Raw Data'
```

```python
df1 = pd.read_csv('Raw Data/RAW_recipes.csv', usecols=['id','nutrition']).sort_values(by=['id'])
df1
```

# Add new columns

```python
cols = ['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat', 'carbohydrates']
```

```python
df = df1.reindex(columns=[*df1.columns.tolist(), *cols], fill_value=0)
df
```

# Dvide (nutrition) column values into seperated columns

```python
for i in range(len(df)):

    #split food nutrition 
    food_nut = df.iloc[i,1]
    nutrition = [float(s) for s in re.findall(r'-?\d+\.?\d*', food_nut)]
    df.iloc[i,2] = nutrition[0]
    df.iloc[i,3] = nutrition[1]
    df.iloc[i,4] = nutrition[2]
    df.iloc[i,5] = nutrition[3]
    df.iloc[i,6] = nutrition[4]
    df.iloc[i,7] = nutrition[5]
    df.iloc[i,8] = nutrition[6]
    
```

# Remove unnecessery columns

```python
df2 = df.drop(['nutrition'], axis=1).set_index('id')
df2
```

# Save the outputs

```python
df2.to_csv('recipes_nutrition.csv')
```
